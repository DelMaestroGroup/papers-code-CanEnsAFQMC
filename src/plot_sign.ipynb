### Sign

```julia
using DelimitedFiles, Plots, LaTeXStrings
```

```julia
ce_data = readdlm("../data/sign_ce_Lx6Ly6_U4.dat")
gce_data = readdlm("../data/sign_gce_Lx6Ly6_U4.dat")

n̄_ce = ce_data[2:14, 1]
sgn_ce = ce_data[2:14, 2]
sgn_ce_err = ce_data[2:14, 3]

n̄_gce = gce_data[2:end, 1]
sgn_gce = gce_data[2:end, 2]
sgn_gce_err = gce_data[2:end, 3];
```

```julia
### Plot
cur_colors = get_color_palette(:auto, plot_color(:white))

plt = plot(
    n̄_ce, sgn_ce, yerror = sgn_ce_err,
    label=L"\mathrm{CE}",
    xlabel = L"\mathrm{Filling}, \langle n \rangle", ylabel=L"\langle \mathrm{Sign} \rangle",
    lc=cur_colors[1],
    mc=cur_colors[1],
    msc=cur_colors[1], msw=1.5,
    marker=(:square, 4), 
    ylim=(-0.05,1.02), xminorticks=3, yminorticks=2,
    fontfamily="Computer Modern",
    legendfontsize = 10,
    guidefontsize = 12,
    tickfontsize = 10,
    framestyle=:box, grid=:false,
    legend=:bottomleft,
    foreground_color_legend = nothing
)

scatter!(
    n̄_ce, sgn_ce,
    lc=cur_colors[1],
    mc=cur_colors[1],
    msc=:black, msw=1.5,
    marker=(:square, 4), 
    label=""
)

plot!(
    plt,
    n̄_gce, sgn_gce, yerror = sgn_gce_err,
    label=L"\mathrm{GCE}",
    lc=cur_colors[2],
    mc=cur_colors[2],
    msc=cur_colors[2], msw=1.5,
    marker=(:square, 4)
)

scatter!(
    n̄_gce, sgn_gce,
    lc=cur_colors[2],
    mc=cur_colors[2],
    msc=:black, msw=1.5,
    marker=(:square, 4), 
    label=""
)

display(MIME"image/png"(), plt)
```
