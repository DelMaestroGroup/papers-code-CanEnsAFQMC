### Energy

```julia
using Plots, DelimitedFiles, LaTeXStrings
```

```julia
data = readdlm("../data/energy_Lx6Ly6_U4.dat");
Î² = data[2:end, 1]

Ece = data[2:end, 2]
Ece_error = data[2:end, 3]

Egce = data[2:end, 4]
Egce_error = data[2:end, 5];
```

```julia
### Plot
cur_colors = get_color_palette(:auto, plot_color(:white))
plt = plot(
    Î², Ece, yerror = Ece_error,
    label=L"\mathrm{CE}",
    xlabel = L"\beta", ylabel=L"E/\mathrm{N}",
    lc=cur_colors[1], mc=cur_colors[1],
    marker=(:square, 4), 
    msw=1.5,
    msc=cur_colors[1],
    xlim=(0.0,13), ylim=(-0.9,-0.1), xminorticks=3, yminorticks=2,
    fontfamily="Computer Modern",
    legendfontsize = 10,
    guidefontsize = 11,
    tickfontsize = 10,
    framestyle=:box, grid=:false,
    legend=:bottomleft,
    foreground_color_legend = nothing
)

scatter!(
    plt,
    Î², Ece, yerror = Ece_error,
    label="",
    mc=cur_colors[1],
    marker=(:square, 4), 
    msw=1.5,
    msc=:black
)

plot!(
    plt,
    Î², Egce,
    label=L"\mathrm{GCE}",
    marker=(:square, 4),
    msw=1.5,
    lc=cur_colors[2], mc=cur_colors[2],
    msc=cur_colors[2]
)

scatter!(
    plt,
    Î², Egce,
    label="",
    mc=cur_colors[2],
    marker=(:square, 4), 
    msw=1.5,
    msc=:black
)

# inset
plot!(
    plt,
    Î²[6:end], Ece[6:end], yerror = Ece_error[6:end],
    xlabel = L"\beta", ylabel=L"E/\mathrm{N}",
    xticks = ([6, 8, 10, 12]),
    lc=cur_colors[1], mc=cur_colors[1],
    label="", marker=(:square, 4),
    msw=1.5,
    msc=cur_colors[1],
    ylim=(-0.86, -0.84),
    inset=bbox(0.50, 0.12, 0.40, 0.40),
    guidefontsize = 8,
    framestyle=:box, grid=:false,
    subplot=2
)

scatter!(
    plt[2],
    Î²[6:end], Ece[6:end],
    label="",
    mc=cur_colors[1],
    marker=(:square, 4), 
    msw=1.5,
    msc=:black
)

plot!(
    plt[2],
    Î²[6:end], Egce[6:end], yerror = Egce_error[6:end],
    lc=cur_colors[2], mc=cur_colors[2],
    msw=1.5,
    label="", marker=(:square, 4),
    msc=cur_colors[2]
)

scatter!(
    plt[2],
    Î²[6:end], Egce[6:end],
    label="",
    mc=cur_colors[2],
    marker=(:square, 4), 
    msw=1.5,
    msc=:black
)

display(MIME"image/png"(), plt)
```
